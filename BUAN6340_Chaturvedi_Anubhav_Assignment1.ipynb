### Q10 Top 5 highest OBP (on base percentage) with at least 500 PAs in 1977?  (Dataframe)

```python
h = df.groupby(['yearID', 'playerID', 'nameFirst', 'nameLast'], as_index = False).sum()
h['OBP'] = (h['H'] + h['BB'] + h['IBB'] + h['SH'] + h['SF']) / h['AB']
h['PA'] = h['AB'] + h['BB'] + h['IBB'] + h['SH'] + h['SF']
h = h[(h['yearID'] == 1977) & (h['PA'] >= 500)]
h.sort_values(by=['OBP'], ascending=[False]).head().reset_index(drop=True)
```

### Q11 Top 8 highest averages in 2013 with at least 300 PAs? (Dataframe)

### Q13 Who led MLB with the most hits the most times?  And how many times?  (Dataframe, Number)

### Q17 Please filter out and show me the top 3 average seasons by Wade Boggs during his career in seasons in which he had at least 500 ABs.  I would like a dataframe sorted by average.

### Q21 Who had the highest OBP is 1986 with at least 400 PAs? (Dataframe)

```python
h = df.groupby(['yearID', 'playerID', 'nameFirst', 'nameLast'], as_index = False).sum()
h['OBP'] = (h['H'] + h['BB'] + h['IBB'] + h['SH'] + h['SF']) / h['AB']
h['PA'] = h['AB'] + h['BB'] + h['IBB'] + h['SH'] + h['SF']
s = h.groupby(['yearID', 'playerID', 'nameFirst', 'nameLast'], as_index = False)[['PA','OBP']].sum()
s[(s.yearID == 1986) & (s.PA > 399)].sort_values(by=['OBP'], ascending=[False]).head(1)
```

### Q22 Same question but for 1997 and only in the NL (check league ID)? (Dataframe)

### Q23 Who had more than the league average HRs in 2012 (filter out all players with less 500 PAs)? (Dataframe)

```python
df1= df[df['yearID']==2012].groupby(['playerID','nameFirst','nameLast','yearID'],as_index=False ).sum()
df1['PA']= df1['AB']+df1['BB']+df1['IBB']+df1['SF']+df1['SH']
df1[(df1['PA']>=500) & (df1['HR']> df1['HR'].mean())][['playerID','nameFirst','nameLast','yearID','HR']].sort_values('HR',ascending=False).reset_index(drop=True)
```

### Q24 Who is the youngest player to hit 50 or more HRs in a single season? (Dataframe)

```python
v= df.groupby(['yearID', 'birthYear', 'playerID', 'nameFirst', 'nameLast'], as_index = False).sum()
v['Age'] = v['yearID'] - v['birthYear']
v[v['HR'] >= 50].sort_values(by = 'Age', ascending = True).head(1)
```

### Q25 Who are the five youngest players to hit 300 or more HRs for their career? (Dataframe)

```python
w = df.groupby(['playerID', 'nameFirst', 'nameLast', 'birthYear', 'yearID']).sum().groupby(level=0).cumsum().reset_index()
w = w[w['HR'] >= 300].sort_values(by = ['yearID', 'birthYear', 'playerID'])[['playerID', 'nameFirst', 'nameLast', 'birthYear', 'yearID', 'HR']]
w['Age'] = w['yearID'] - w['birthYear']
w.sort_values(by = 'Age', ascending = True).head()
```

### (2) Using a line graph please graph the average HRs per AB (think about this) per season

```python
plt.style.use('seaborn-bright')
df1 = df[df['HR']>=0].groupby('yearID').mean()['HR']
df2 = df[df['AB']>=0].groupby('yearID').mean()['AB']
df3 = df1/df2
df3.plot.line(x=df1.index,y=df2,figsize=(15,4),lw=1)
plt.ylabel('Average HRs per AB')
plt.xlabel('Years')
plt.title('Year v/s Average, Home Runs (HRs) per At-bat(AB)')
plt.show()
```
