### Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Importing Dataset

```python
df =  pd.read_csv('Batting.csv')
df.columns
```

```python
df.shape
```

### Visualization 1 - Home Runs hit every year

```python
k = df.groupby(['yearID'],as_index=False)[['HR']].sum()
```

```python
plt.plot(k['yearID'], k['HR'], 'blue')
plt.xlabel('Year')
plt.ylabel('Home Runs Hit')
plt.title('Number of HRs hit over the years')
plt.style.use('fivethirtyeight')
```

```python
plt.show()
```

### Visualization 2 - Wage Boggs Average over the years

```python
i = df[(df.nameFirst == 'Wade') & (df.nameLast == 'Boggs')]
i
Average = i['H']/i['AB']
```

```python
plt.plot(i['yearID'], Average, 'green')
plt.xlabel('Year')
plt.ylabel('Wage Boggs Average')
plt.title('Wage Boggs Average over his career')
plt.style.use('ggplot')
```

```python
plt.show()
```

### Visualization 3 - Barry Bonds Home Runs over the years

```python
a = df[(df.nameFirst == 'Barry') & (df.nameLast == 'Bonds')]
a
```

```python
plt.plot(a['yearID'], a['HR'], 'red')
plt.xlabel('Year')
plt.ylabel('Home Runs hit')
plt.title('Barry Bonds Home Runs over the years')
plt.style.use('ggplot')
plt.show()
```

### Visualization 4  - How different IgIDs compare in terms of Home Runs

```python
c = df.groupby('lgID',as_index=False).sum()
c
```

```python
sns.barplot(x='lgID',y='HR',data=c)
plt.xlabel('IgID')
plt.ylabel('Home Runs hit')
plt.title('How different IgIDs compare in terms of HR')
plt.show()
```
